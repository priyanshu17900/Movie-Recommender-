```python
import re

def clean_tokens(words):
    cleaned = []
    for w in words:
        w = w.lower()
        w = re.sub(r'[^a-z0-9]', '', w)  # remove punctuation
        if w:  # drop empty strings
            cleaned.append(w)
    return cleaned
```

```python
movies['overview'] = movies['overview'].apply(clean_tokens)
movies['genres'] = movies['genres'].apply(clean_tokens)
movies['keywords'] = movies['keywords'].apply(clean_tokens)
movies['tagline'] = movies['tagline'].apply(clean_tokens)
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres'] * 4
    + movies['keywords'] * 4
    + movies['tagline'] * 2
)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem_words(words):
    return " ".join(ps.stem(w) for w in words)

movies['tags'] = movies['tags'].apply(stem_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=6000,
    stop_words='english'
)
movies = movies.sample(5000, random_state=42).reset_index(drop=True)

vectors = tfidf.fit_transform(movies['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(movie):
    idx = movies[movies['title'] == movie].index[0]
    scores = cosine_similarity(vectors[idx], vectors).flatten()
    top = scores.argsort()[::-1][1:6]

    for i in top:
        print(movies.iloc[i].title)
```
